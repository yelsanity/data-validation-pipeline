```python
import pandas as pd
item_orders_df = pd.read_csv('../csv-files/olist_order_items_dataset.csv')
item_orders_df.head()
```

```python
#checking character lengths for IDs
order_id_len = item_orders_df['order_id'].str.len()
product_id_len = item_orders_df['order_id'].str.len()
seller_id_len = item_orders_df['order_id'].str.len()

standard_order_id_len = order_id_len.mode()[0]
standard_product_id_len = product_id_len.mode()[0]
standard_seller_id_len = seller_id_len.mode()[0]

invalid_order_id_len = order_id_len != standard_order_id_len
invalid_product_id_len = order_id_len != standard_product_id_len
invalid_seller_id_len = order_id_len != standard_seller_id_len

invalid_order_id_rows = item_orders_df[invalid_order_id_len]
invalid_product_id_rows = item_orders_df[invalid_product_id_len]
invalid_seller_id_rows = item_orders_df[invalid_seller_id_len]

print("order_id standard length:", standard_order_id_len)
print("order_id rows with invalid lenght:", len(invalid_order_id_rows))
print("product_id standard length:", standard_product_id_len)
print("product_id rows with invalid lenght:", len(invalid_product_id_rows))
print("seller_id standard length:", standard_seller_id_len)
print("seller_id rows with invalid lenght:", len(invalid_seller_id_rows))
```

```python
unique_order_id = item_orders_df['order_id'].nunique()
print("order_id unique:", unique_order_id)
print("order_id rows:", len(item_orders_df))
```

```python
dups = item_orders_df.duplicated()
dup_count = dups.sum()
print(dup_count)
```

```python
# Find rows with the same order_id
duplicate_order_ids = item_orders_df[item_orders_df.duplicated('order_id', keep=False)]

# Sort by order_id to see the items grouped together
print(duplicate_order_ids[['order_id', 'order_item_id', 'product_id', 'price']].sort_values('order_id').head(4))
```

```python
# Group by order and product to count the total quantity
order_quantities = item_orders_df.groupby(['order_id', 'product_id']).size().reset_index(name='quantity')

# View orders where someone bought multiples of the same item
multi_quantities = order_quantities[order_quantities['quantity'] > 1]
print(multi_quantities.head())
```
